## manipulating values in a dataframe

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
titanic.loc[1, 'age'] = 40
```

```python
titanic.head()
```

```python
titanic.iloc[1, 3] = 41
```

```python
titanic
```

```python
titanic.loc[1:3, 'age'] = 42
```

```python
titanic
```

```python
titanic.iloc[1:4, 3] = [43, 44, 45]
```

```python
titanic
```

```python
index_babies = titanic.loc[titanic.age < 1, 'age'].index
```

```python
titanic.loc[titanic.age < 1, 'age'] = 1
```

```python
titanic.loc[index_babies]
```

```python
titanic
```

```python
titanic.loc[0, 'survived': 'sex'] = [1, 1, 'female']
```

```python
titanic
```

```python
titanic.replace(0, 'Zero')
```

## view vs. copy

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
age = titanic.age
```

```python
age.head()
```

```python
age._is_view
```

```python
age._is_copy is None
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic
```

```python
df_baby = titanic[titanic.age <1]
```

```python
df_baby._is_view
```

```python
df_baby._is_copy is None
```

```python
df_baby._is_copy()
```

```python
df_baby.age = 1
```

```python
df_baby
```

```python
titanic.loc[index_babies]
```

## if you want to work with and manipulate the whole dataframe avoid chained indexing

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
titanic.iloc[1, 3] = 40
```

```python
titanic.head()
```

```python
index_babies = titanic.loc[titanic.age < 1, 'age'].index
```

```python
titanic.loc[titanic.age < 1, 'age'] = 1
```

```python
titanic.loc[index_babies]
```

## if you want to work with and manipulate a slice of a dataframe and make a copy with .copy()

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
age = titanic.age.copy()
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

```python
baby_ages = titanic.loc[titanic.age < 1, ['age', 'sex']].copy()
```

```python
baby_ages
```

```python
baby_ages['age'] = 1
```

```python
baby_ages
```

```python
titanic.loc[index_babies]
```
