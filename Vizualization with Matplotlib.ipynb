## the plot() method

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot(subplots=True, figsize=(15,12), sharex=False)
plt.show()
```

```python
titanic.age.plot(figsize=(12,8))
plt.show()
```

## customization

```python
xticks=[x for x in range(0,901,50)]
xticks
```

```python
yticks = [y for y in range(0, 81, 5)]
yticks
```

```python
titanic.age.plot(figsize=(12,8), fontsize=13, c='r', linestyle='-', xticks=xticks, yticks=yticks, rot=45)
plt.title('Titanic - Ages', fontsize=20)
plt.legend(loc='lower left', fontsize=15)
plt.xlabel('Passenger No', fontsize=15)
plt.ylabel('Age', fontsize=15)
plt.grid()
plt.show()
```

## histograms

```python
plt.style.use('seaborn')
```

```python
titanic
```

```python
titanic.age.value_counts(bins=10)
```

```python
titanic.age.plot(kind='hist', figsize=(12,8), fontsize=15, bins=80)
plt.show()
```

```python
titanic.age.hist(figsize=(12,8), bins=80, xlabelsize=15, ylabelsize=15)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.hist(titanic.age, bins=80, density=True, cumulative=True)
plt.show()
```
