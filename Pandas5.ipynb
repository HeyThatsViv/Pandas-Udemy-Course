## the agg() method

```python
titanic
```

```python
titanic.describe()
```

```python
titanic.agg(['mean', 'std'])
```

```python
titanic.agg(['mean', 'std', 'min', 'max', 'median'])
```

```python
titanic.agg({'survived': 'mean', 'age':['min', 'max']})
```

```python
titanic.loc[:, ['survived','pclass','fare']]
```

```python
titanic.nlargest(10, ['fare', 'age'])
```

```python
children = titanic.age < 16
parents = titanic.parch == 0

titanic.loc[children & parents, ['age', 'sex', 'survived']].sort_values('age')
```

```python
females = titanic.sex == 'female'
lower_class = titanic.pclass != 1
col_list = ['pclass', 'age', 'fare']
slice = titanic.loc[females & lower_class, col_list]
slice.nlargest(20, ['age', 'fare'])
```

```python
children = titanic.age < 18
low_class = titanic.pclass == 3
survival = titanic.survived == 1
slice = titanic.loc[children & low_class & survival, :]
slice.sort_values(['age', 'fare'], ascending=[True, False])
```

```python
children = titanic.age < 18
low_class = titanic.pclass == 3
survival = titanic.survived == 1
df_slice = titanic.loc[children & low_class & survival, :]
df_slice.sort_values(['age', 'fare'], ascending=[True, False])
```

```python
titanic.age.agg('mean')
```
