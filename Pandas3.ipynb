## filtering dataframes with many conditions (AND)

```python
titanic
```

```python
mask1 = titanic.sex == 'male'
mask1
```

```python
mask2 = titanic.age > 14
mask2
```

```python
(mask1 & mask2).head()
```

```python
male_surv = titanic.loc[mask1 & mask2, ['survived', 'pclass', 'sex', 'age']]
```

```python
male_surv
```

```python
male_surv.info()
```

```python
male_surv.describe()
```

```python
titanic.describe()
```

```python
wom_child_surv = titanic.loc[~mask1 & ~mask2, ['survived', 'pclass', 'sex', 'age']]
```

```python
wom_child_surv
```

```python
wom_child_surv.describe()
```

## filtering dataframes with many conditions (OR)

```python
titanic
```

```python
mask1 = titanic.sex == 'female'
mask1
```

```python
mask2 = titanic.age < 14
mask2
```

```python
(mask1 | mask2)
```

```python
titanic.loc[mask1 | mask2]
```

```python
wom_or_chi = titanic.loc[mask1 | mask2, ['survived', 'pclass', 'sex', 'age']]
```

```python
wom_or_chi
```

```python
wom_or_chi.describe()
```

## advanced filtering with between(), isin(), and ~

```python
summer = pd.read_csv('summer.csv')
```

```python
summer
```

```python
og_1988 = summer.loc[summer.Year==1988]
```

```python
og_1988
```

```python
og_1988.info()
```

```python
og_since1992 = summer.loc[summer.Year >= 1992]
```

```python
og_since1992
```

```python
summer.Year.between(1960, 1969).head()
```

```python
og_60s = summer.loc[summer.Year.between(1960, 1969, inclusive=True)]
```

```python
og_60s
```

```python
my_favorite_games = [1972, 1996]
```

```python
summer.Year.isin(my_favorite_games).head()
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_favorite_games)]
```

```python
og_72_96
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_favorite_games)]
```

```python
og_not_72_96.Year.value_counts()
```
