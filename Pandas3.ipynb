## filtering dataframes with one condition

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic
```

```python
titanic.sex.head()
```

```python
titanic.sex == 'male'
```

```python
titanic[titanic.sex == 'male']
```

```python
titanic[titanic.sex == 'male']['fare']
```

```python
titanic.loc[titanic.sex == 'male', 'fare']
```

```python
mask1 = titanic.sex == 'male'
```

```python
titanic_male = titanic.loc[mask1]
```

```python
titanic_male
```

```python
mask2 = titanic.dtypes == object
```

```python
titanic.loc[:, ~mask2]
```

```python
titanic.loc[mask1, ~mask2]
```

```python
titanic.loc[mask1, mask2]
```

## filtering dataframes with many conditions (AND)

```python
titanic
```

```python
mask1 = titanic.sex == 'male'
mask1
```

```python
mask2 = titanic.age > 14
mask2
```

```python
(mask1 & mask2).head()
```

```python
male_surv = titanic.loc[mask1 & mask2, ['survived', 'pclass', 'sex', 'age']]
```

```python
male_surv
```

```python
male_surv.info()
```

```python
male_surv.describe()
```

```python
titanic.describe()
```

```python
wom_child_surv = titanic.loc[~mask1 & ~mask2, ['survived', 'pclass', 'sex', 'age']]
```

```python
wom_child_surv
```

```python
wom_child_surv.describe()
```

## filtering dataframes with many conditions (OR)

```python
titanic
```

```python
mask1 = titanic.sex == 'female'
mask1
```

```python
mask2 = titanic.age < 14
mask2
```

```python
(mask1 | mask2)
```

```python
titanic.loc[mask1 | mask2]
```

```python
wom_or_chi = titanic.loc[mask1 | mask2, ['survived', 'pclass', 'sex', 'age']]
```

```python
wom_or_chi
```

```python
wom_or_chi.describe()
```
